```python
import string

import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from wordcloud import WordCloud

%matplotlib inline
```

```python
with open('bitcoin_whitepaper.md', 'r') as f:
    text = f.read()
```

```python
# following lines need to be executed only once

# import nltk
# nltk.download('stopwords')

stop_words = stopwords.words('english')
```

```python
words = text.split()
cleaned = []

for w in words:
    # remove punctuation
    w = w.translate(str.maketrans('', '', string.punctuation)).lower()
    # remove stopwords, digits, and single chars
    if w not in stop_words and not w.isdigit() and len(w) > 1:
        cleaned.append(w)
```

```python
wordcloud = WordCloud(relative_scaling=.5, width=1200, height=800, background_color='black', colormap="Oranges",
                      max_words=300).generate_from_text(' '.join(cleaned))
plt.figure(figsize=(15, 15), dpi=300)
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('bitcoin_whitepaper_wordcloud.png', dpi=300)
```
